```python
def get_stimuli_where_no_behavior(traces, start, stop, num_traces_to_return, stim_times):
    amount_of_behavior = np.mean(np.abs(traces[:,start:stop]),axis=-1)
    indicies = np.argsort(amount_of_behavior)
    top_x_indicies = indicies[:num_traces_to_return]
    return traces[top_x_indicies,:], np.asarray(stim_times)[top_x_indicies]
```

```python
###########################
### Create Notch Filter ###
###########################

fs = 10.0  # Sample frequency (Hz)
f0 = 1.8797  # Frequency to be removed from signal (Hz)
Q = .8  # Quality factor <---- IMPORTANT
w0 = f0/(fs/2)  # Normalized Frequency

# Design notch filter
b_notch, a_notch = signal.iirnotch(w0, Q)

# Frequency response

w, h = signal.freqz(b_notch, a_notch)

# Generate frequency axis
freq = w*fs/(2*np.pi)
```

```python
def make_STA_brain(neural_signals, neural_timestamps, event_times_list, neural_bins):
    #### super voxel version
    
    num_z = neural_signals.shape[0]

    STA_brain = []
    for z in range(num_z):
        all_bin_indicies = []
        for stim_idx in range(len(event_times_list)):
            stim_time = event_times_list[stim_idx]
            stim_centered_bins = neural_bins + stim_time
            bin_indicies = np.digitize(neural_timestamps[:,z] , stim_centered_bins)
            all_bin_indicies.append(bin_indicies)
        all_bin_indicies = np.asarray(all_bin_indicies)

        avg_neural_across_bins = []
        for bin_num in np.arange(1,len(neural_bins)):
            this_bin_sample_times = list(np.where(all_bin_indicies==bin_num)[1])
            average_neural_in_bin = np.mean(neural_signals[z,:,this_bin_sample_times],axis=0)
            avg_neural_across_bins.append(average_neural_in_bin)
        avg_neural_across_bins = np.asarray(avg_neural_across_bins)
        STA_brain.append(avg_neural_across_bins)
    STA_brain = np.asarray(STA_brain)
    return STA_brain
```

```python
# Example setup - Ensure these variables are defined as per your data
# roi_signal: Dictionary containing the signals for different conditions
# roi_ids: Dictionary containing the ROI identifiers for each hemisphere

def calculate_z_score_diff(hemi):
    # Ensure that ROI indices are within the range of available signal data
    valid_rois = [roi for roi in roi_ids[hemi] if roi in roi_signal['mean'][hemi]['ve_180'] and roi in roi_signal['mean'][hemi]['ve_0']]
    
    # Calculate z-scores for each condition using the validated ROIs
    z_scores_ve_180 = stats.zscore(np.array([roi_signal['mean'][hemi]['ve_180'][roi] for roi in valid_rois]), axis=1)
    z_scores_ve_0 = stats.zscore(np.array([roi_signal['mean'][hemi]['ve_0'][roi] for roi in valid_rois]), axis=1)
    
    # Compute the difference between the z-scores
    z_score_diff = z_scores_ve_180 - z_scores_ve_0

    return z_score_diff, valid_rois

# Calculate z-score differences for each hemisphere and plot heatmaps
for hemi in ['left', 'right']:
    z_score_diff, valid_rois = calculate_z_score_diff(hemi)
    # Assume times is already defined as an array of time points
    roi_names = [roi_ids[hemi][roi] for roi in roi_ids[hemi]]  # Get the ROI names in the original order

    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(z_score_diff, cmap='bwr', center=0, yticklabels=roi_names)
    ax.set_title(f'Z-Score Difference Heatmap ({hemi} Hemisphere): ve_180 - ve_0')
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('ROI')
    plt.xticks(ticks=np.linspace(0, len(times)-1, num=5), labels=[f"{int(time)} ms" for time in np.linspace(times[0], times[-1], num=5)])
    plt.show()
```

```python
def extract_roi_signal_traces(roi_ids, roi_masks, all_warps, condition, hemi, signal_type):
    t0 = time.time()
    roi_time_avgs = []
    for roi in roi_ids[hemi]:
        mask = roi_masks[roi]
        masked_data = all_warps[condition][:,:,:,::-1]*mask[np.newaxis,:,:,:] #note z-flip
        if signal_type == 'max':
            roi_time_avg = np.max(masked_data,axis=(1,2,3))
        elif signal_type == 'mean':
            roi_time_avg = np.mean(masked_data,axis=(1,2,3))
        roi_time_avgs.append(roi_time_avg)
    print(time.time()-t0)
    return np.asarray(roi_time_avgs)
```
