```python
def STA_supervoxel_to_full_res(STA_brain, cluster_labels, dim_x=256, dim_y=128, dim_z=49):
    n_clusters = STA_brain.shape[2]
    n_tp = STA_brain.shape[1]
    
    reformed_STA_brain = []
    for z in range(dim_z):
        colored_by_betas = np.zeros((n_tp, dim_x*dim_y))
        for cluster_num in range(n_clusters):
            cluster_indicies = np.where(cluster_labels[z,:]==cluster_num)[0]
            colored_by_betas[:,cluster_indicies] = STA_brain[z,:,cluster_num,np.newaxis]
        colored_by_betas = colored_by_betas.reshape(n_tp,dim_x,dim_y)
        reformed_STA_brain.append(colored_by_betas)
    return np.asarray(reformed_STA_brain)
```

```python
def load_batches(X, batch_size):
    """Yield successive batches from X."""
    for i in range(0, X.shape[0], batch_size):
        yield X[i:i + batch_size]
```

```python
#########################
#### IncrementalPCA #####
#########################
n_components = 50  # for example, number of components to keep
batch_size = 7000  # chosen based on memory calculation

ipca = IncrementalPCA(n_components=n_components)

# Loop to load and process batches
for X_batch in load_batches(X, batch_size):  # Define or use your own batch loading function
    ipca.partial_fit(X_batch)
```

```python
n_components = 50  # for example, number of components to keep
batch_size = 7000  # chosen based on memory calculation

ipca = IncrementalPCA(n_components=n_components)

# Assuming you can determine the total number of batches beforehand
total_batches = (X.shape[0] + batch_size - 1) // batch_size  # This calculates the number of batches
print(f"Total number of batches: {total_batches}")

# Loop to load and process batches
batch_count = 0
for X_batch in load_batches(X, batch_size):
    ipca.partial_fit(X_batch)
    batch_count += 1
    print(f"Processed batch {batch_count}/{total_batches}")

print("Incremental PCA fitting complete.")

# If you want, you can also print some information about the components
print("Principal components shape:", ipca.components_.shape)
print("Explained variance by each component:", ipca.explained_variance_)
```

```python
# After all data is processed, access the components
principal_components = ipca.components_  # These are your eigenvectors
explained_variance_ratios = ipca.explained_variance_ratio_
eigenvalues = ipca.explained_variance_  # These approximate the eigenvalues
```
